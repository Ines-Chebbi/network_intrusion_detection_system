```python
@app.route('/predict', methods=['POST'])
def predict():
    try:
        # Extract form data
        protocol_type= request.form.get('protocol_type')
        service = request.form.get('service')
        flag = request.form.get('flag')
        src_bytes = request.form.get('src_bytes')
        dst_bytes = request.form.get('dst_bytes')
        count = request.form.get('count')
        same_srv_rate = request.form.get('same_srv_rate')
        diff_srv_rate = request.form.get('diff_srv_rate')
        dst_host_srv_count = request.form.get('dst_host_srv_count')
        dst_host_same_srv_rate= request.form.get('dst_host_same_srv_rate')


        # Debug print to verify form data
        print("Form Data Received:", request.form)

        # Process form data and prepare for prediction
        input_data = pd.DataFrame({
            'protocol_type': [protocol_type],
            'service': [service],
            'flag': [flag],
            'src_bytes': [src_bytes],
            'dst_bytes': [dst_bytes],
            'count': [count],
            'same_srv_rate': [same_srv_rate],
            'diff_srv_rate': [diff_srv_rate],
            'dst_host_srv_count': [dst_host_srv_count],
            'dst_host_same_srv_rate': [dst_host_same_srv_rate]

          })

        # Convert categorical features to numerical values
        input_data['protocol_type'] = input_data['protocol_type'].map({'icmp': 0, 'tcp': 1,'udp':2})

        input_data['service'] = input_data['service'].map({'http': 0, 'private': 1,'domain_u':2 ,'smtp': 3,
                                              'ftp_data': 4,'exec': 5,'nnsp': 6,'finger': 7,'http_8001': 8,'tim_i': 9,'pm_dump': 10
                                                           ,'urh_i': 11, 'red_i': 12,'other': 13

                    })

        input_data['flag'] = input_data['flag'].map({'OTH': 0, 'REJ': 1,'RSTO':2 ,'RSTOS0': 3,
                                              'RSTR': 4,'S0': 5,'S1': 6,'S2': 7,'S3': 8,'SF': 9,'SH': 10
                                                           })


        # Convert tenure, MonthlyCharges, and TotalCharges to numeric
        input_data['src_bytes'] = pd.to_numeric(input_data['src_bytes'], errors='coerce')
        input_data['dst_bytes'] = pd.to_numeric(input_data['dst_bytes'], errors='coerce')
        input_data['count'] = pd.to_numeric(input_data['count'], errors='coerce')
        input_data['same_srv_rate'] = pd.to_numeric(input_data['same_srv_rate'], errors='coerce')
        input_data['diff_srv_rate'] = pd.to_numeric(input_data['diff_srv_rate'], errors='coerce')
        input_data['dst_host_srv_count'] = pd.to_numeric(input_data['dst_host_srv_count'], errors='coerce')
        input_data['dst_host_same_srv_rate'] = pd.to_numeric(input_data['dst_host_same_srv_rate'], errors='coerce')

        # Ensure columns match the training data
        columns = ['protocol_type', 'service', 'flag', 'src_bytes', 'dst_bytes', 'count',
            'same_srv_rate', 'diff_srv_rate', 'dst_host_srv_count', 'dst_host_same_srv_rate']

        input_data = input_data[columns]

        # Drop the feature names (convert to numpy array)
        input_array = input_data.to_numpy()

        # Debug prints to ensure data transformation is correct
        print("Processed Input DataFrame:\n", input_data)
        print("Processed Input Array:\n", input_array)

        # Make predictions
        prediction = model.predict(input_array)[0]
        prediction_prob = model.predict_proba(input_array)[0]

        # Debug print to check prediction results
        print("Prediction:", prediction)
        print("Prediction Probability:", prediction_prob)

        # Return the result
        return jsonify({'prediction': 'This network traffic is likely to be normal.' if prediction == 1 else 'This network traffic is likely to be an intrusion.'
                        })

    except Exception as e:
        print("Error during prediction:", e)
        return jsonify({'error': str(e)})
```
